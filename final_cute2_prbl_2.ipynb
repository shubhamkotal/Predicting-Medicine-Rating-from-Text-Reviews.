```python
import pandas as pd
train=pd.read_csv("traindata-1548399384664.csv",encoding='latin-1')
train
```

```python
test=pd.read_csv("testdata-1548399401363.csv",encoding='latin-1')
test
```

```python
train=train.drop('Unnamed: 0',axis=1)
test=test.drop('Unnamed: 0',axis=1)
test.shape
train.shape
```

```python
test['Rating'].value_counts()
```

```python
#joining all for tokenization
train["var1"] = train["effectiveness"]+"."+train["sideEffects"]+"."+train["condition"]+"."+ train["benefitsReview"]+"."+train["sideEffectsReview"]+"."+train["commentsReview"]
```

```python
test["var2"] = test["effectiveness"]+"."+test["sideEffects"]+"."+test["condition"]+"."+ test["benefitsReview"]+"."+test["sideEffectsReview"]+"."+test["commentsReview"]
```

```python
## Define the sequence lengths, max number of words
MAX_SEQUENCE_LENGTH = 300  
MAX_NB_WORDS = 20000        # Top 20000 frequently occuring words
```

```python
## Import Libraries 
import numpy as np      # for array operations
import pandas as pd     # for reading data operations

from keras.preprocessing.text import Tokenizer          # for tokenizing text
from keras.preprocessing.sequence import pad_sequences  # for padding sentences with zeros. To make the sentence length same
from keras.utils import to_categorical                  # for one-hot encoding of the labels
from keras.layers import Dense, Input, Flatten, Dropout, BatchNormalization
from keras.layers import Conv1D, MaxPooling1D, GlobalMaxPool1D, Embedding
from keras.models import Sequential
```

```python
#convert to string
train["var1"] = train["var1"].astype("str")
test["var2"] = test["var2"].astype("str")
```

```python
tokenizer = Tokenizer(num_words=MAX_NB_WORDS)   # get the frequently occuring words
tokenizer.fit_on_texts(train.var1)           
train_sequences = tokenizer.texts_to_sequences(train.var1)
test_sequences = tokenizer.texts_to_sequences(test.var2)

word_index = tokenizer.word_index               # dictionary containing words and their index
#print(tokenizer.word_index)                   # print to check
print('Found %s unique tokens.' % len(word_index)) # total words in the corpus
train_data = pad_sequences(train_sequences, maxlen=MAX_SEQUENCE_LENGTH) # get only the top frequent words on train
test_data = pad_sequences(test_sequences, maxlen=MAX_SEQUENCE_LENGTH)   # get only the top frequent words on test

print(train_data.shape)
print(test_data.shape)
```

```python
train_labels = train["rating"]
test_labels = test["Rating"]

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()                  # converts the character array to numeric array. Assigns levels to unique labels.
le.fit(train_labels)
train_labels = le.transform(train_labels)
test_labels = le.transform(test_labels)

#print(le.classes_)
print(np.unique(train_labels, return_counts=True))
print(np.unique(test_labels, return_counts=True))
```

```python
labels_train = to_categorical(np.asarray(train_labels))
labels_test = to_categorical(np.asarray(test_labels))
print('Shape of data tensor:', train_data.shape)
print('Shape of label tensor:', labels_train.shape)
print('Shape of label tensor:', labels_test.shape)
```

```python
from keras.regularizers import l2
from keras.layers import Dense, Input, LSTM, ActivityRegularization


print('Training model.')
EMBEDDING_DIM = 100

model1 = Sequential()
model1.add(Embedding(MAX_NB_WORDS,
                    EMBEDDING_DIM,
                    input_length=MAX_SEQUENCE_LENGTH
                    ))
model1.add(Conv1D(32, 3, activation='relu'))
model1.add(Conv1D(32, 3, activation='relu'))
model1.add(GlobalMaxPool1D())
model1.add(Dropout(0.3))
#ActivityRegularization(l1=0.001, l2=0.001)

model1.add(Dense(512, activation='relu'))
model1.add(Dense(3, activation='softmax'))
```

```python
model1.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model1.summary()
```

```python
model1.fit(train_data, labels_train,
          batch_size=32,
          epochs=15,
          validation_split=0.3)
```

```python
model1.evaluate(test_data, labels_test)
```

```python
# Building an LSTM model

EMBEDDING_DIM = 100

model2 = Sequential() # Call Sequential to initialize a network
model2.add(Embedding(input_dim = MAX_NB_WORDS, 
                    input_length = MAX_SEQUENCE_LENGTH, 
                    output_dim = EMBEDDING_DIM)) # Add an embedding layer which represents each unique token as a vector
model2.add(LSTM(32, return_sequences=True)) # Add an LSTM layer
model2.add(LSTM(64, return_sequences=False))
model2.add(Dense(3, activation='softmax')) # Add an ouput layer. Since classification, 3 nodes for 3 classes.
```

```python
model2.summary()
```

```python
from keras.optimizers import Adam
adam = Adam(lr=0.001)
```

```python
model2.compile(optimizer=adam,                  # 'Adam' is a variant of gradient descent technique
              loss='categorical_crossentropy', # categorical_crossentropy for multi-class classification
              metrics=['accuracy'])            # These metrics are computed for evaluating and stored in history

model2.fit(train_data, labels_train, epochs=15,batch_size=32, validation_split=0.3)


```

```python
model2.evaluate(test_data, labels_test)
```

```python
model3.evaluate(test_data, labels_test)
```

```python
#CNN and LSTM

model4 = Sequential()
model4.add(Embedding(MAX_NB_WORDS,
                    EMBEDDING_DIM,
                    input_length=MAX_SEQUENCE_LENGTH
                    ))
model4.add(Conv1D(32, 3, activation='relu'))
model4.add(MaxPooling1D(pool_size=2))
model4.add(LSTM(64))
model4.add(Dense(3, activation='sigmoid'))
```

```python
#class imbalance


from sklearn.utils import resample



high= train[train.rating=='high']
medium= train[train.rating=='medium']
low= train[train.rating=='low']


new_high = resample(high, 
                           replace=True,     # sample with replacement
                           n_samples=1080,    # to match majority class
                           random_state=123) 



new_medium = resample(medium, 
                           replace=True,     # sample with replacement
                           n_samples=1016,    # to match majority class
                           random_state=123) 

new_low = resample(low, 
                           replace=True,     # sample with replacement
                           n_samples=1011,    # to match majority class
                           random_state=123) 


# Combine back

train2 = pd.concat([new_high, new_medium,new_low])

train2.rating.value_counts()
```
